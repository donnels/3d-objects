```openscad
width=135;
depth=115;
height=2;
holeDist=5;
```

```openscad
// blanking plate4 for back of monitor
module plate(w,d,h) {
    cube([w,d,h]);
}
```

```openscad
%%display plate(width,depth,height);
```

```openscad
module hole() {
    radius=2;
    height=2;
    $fn=360;
    cylinder(r=radius,h=height+1);
}
```

```openscad
%%display hole();
```

```openscad
//
difference() {
    plate(width,depth,height);
    translate([holeDist,holeDist,-.5]) hole();
    translate([width-holeDist,holeDist,-.5]) hole();
    translate([holeDist,depth-holeDist,-.5]) hole();
    translate([width-holeDist,depth-holeDist,-.5]) hole();
};
%display
```
